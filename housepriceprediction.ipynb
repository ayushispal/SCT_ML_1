```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("housepricepredict.csv")
print(data.head())
```

```python
x= data[['sqft_living', 'bedrooms', 'bathrooms']]
y = data['price']
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
model = LinearRegression()
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
print("Mean Squared Error:", mean_squared_error(y_test, y_pred))
print("R² Score:", r2_score(y_test, y_pred))
```

```python
sqft = float(input("Enter square footage: "))
bed = int(input("Enter number of bedrooms: "))
bath = int(input("Enter number of bathrooms: "))

user_input = [[sqft, bed, bath]]
predicted_price = model.predict(user_input)

print(f"Predicted house price: ₹{predicted_price[0]:,.2f}")
```

```python
plt.scatter(y_test, y_pred, color='grey')
plt.xlabel("Actual Price")
plt.ylabel("Predicted Price")
plt.title("Actual vs Predicted House Prices")
plt.grid(True)
plt.show()
```
